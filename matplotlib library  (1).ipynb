# bar plot

```python
import matplotlib.pyplot as plt
x = ["python","c","c++","java"]
y = [85,70,60,82]
plt.xlabel("language",fontsize = 15)
plt.ylabel("number",fontsize = 15)
plt.title("difference",fontsize = 15)
c=["r","b","y","g"]
plt.bar(x,y,width = 0.4,color=c,align ="edge",linewidth = 3,linestyle = ":",alpha=0.6,label="popularity")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python","c","c++","java"]
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel("language",fontsize = 15)
plt.ylabel("number",fontsize = 15)
plt.title("difference",fontsize = 15)


plt.bar(p,y,width,color="r",align ="edge",label="popularity")
plt.bar(p1,z,width,color="y",align ="edge",label="popularity1")

plt.xticks(p+width,x,rotation = 20)
plt.legend()
plt.show()
```

# scatter plot

```python
import matplotlib.pyplot as plt

x = [2,5,4,6,7,8,3,5]
y = [3,5,2,4,1,7,8,5]
y2 = [4,5,1,2,3,6,7,8]
colors=[10,30,55,68,79,80,40,50]
sizes = [99,100,250,200,100,89,110,111]
plt.scatter(x,y,c=colors,s=sizes,marker="*",cmap="Blues")
plt.scatter(x,y2,color="r",s=sizes,marker="*")
t = plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("Scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

# pie plot

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y =["c","c++","python","java"]
ex = [0.3,0.0,0.0,0.0]
c = ["r","m","y","g"]
plt.pie(x,labels=y,explode= ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1,labeldistance=1.1,startangle=90,
        textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":5},rotatelabels=True)

plt.title("pie plot")
plt.legend(loc=2)
plt.show()
```

# stem plot

```python
import matplotlib.pyplot as plt
x= [2,5,7,3,6]
y=[1,4,6,5,3]
plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="y",label = "python",orientation = "horizontal")
plt.legend()
plt.show()
```

# stack plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area = [2,4,5,6,7]
area1 = [4,1,3,6,4]
area2 = [4,3,5,7,6]
l = ["area","area1","area2"]
plt.stackplot(x,area,area1,area2,labels = l,
              colors=["c","m","k"])#baseline = zero,wiggle,sym
plt.title("stack plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2)
plt.show()
```

# step plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,6,4]

plt.step(x,y,marker = "o",ms=10,mfc="c",label = "stepplot")
plt.title("step plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc =2)
plt.grid()
plt.show()
```

# fill_between_plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,3,2,4,5])
y = np.array([2,4,5,2,4])
plt.plot(x,y,color="k")
plt.title("fillbetween")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.fill_between(x,y,color="c",where=(x>=2) & (x<=4))
plt.show()
```
